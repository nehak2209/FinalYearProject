```python
import nltk
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

def preprocess_text(text):
    # Convert to lowercase
    text = text.lower()

    # Remove special characters and digits
    text = re.sub(r'[^a-zA-Z.]', ' ', text)

    # Tokenize the text
    tokens = word_tokenize(text)

    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [token for token in tokens if token not in stop_words]

    # Join the filtered tokens back into a single string
    preprocessed_text = ' '.join(filtered_tokens)

    return preprocessed_text

# Download required NLTK data
nltk.download('punkt')
nltk.download('stopwords')


preprocessed_text = preprocess_text(text)
print(preprocessed_text)
```

```python
import spacy

def perform_ner(text):
    # Load the pre-trained SpaCy model
    nlp = spacy.load("en_core_web_sm")

    # Perform NER on the text
    doc = nlp(text)

    # Initialize lists to store tactics, techniques, and procedures
    tactics = []
    techniques = []
    procedures = []

    # Iterate over the entities and classify them into tactics, techniques, or procedures
    for entity in doc.ents:
        if entity.label_ == "TACTIC":
            tactics.append(entity.text)
        elif entity.label_ == "TECHNIQUE":
            techniques.append(entity.text)
        elif entity.label_ == "PROCEDURE":
            procedures.append(entity.text)

    return tactics, techniques, procedures


tactics, techniques, procedures = perform_ner(preprocessed_text)

print("Tactics:", tactics)
print("Techniques:", techniques)
print("Procedures:", procedures)
```

```python
from transformers import pipeline

def perform_ner(text):
    # Load the pre-trained NER model
    ner_model = pipeline("ner")

    # Perform NER on the text
    entities = ner_model(text)

    # Initialize lists to store tactics, techniques, and procedures
    tactics = []
    techniques = []
    procedures = []

    # Iterate over the entities and classify them into tactics, techniques, or procedures
    for entity in entities:
        if entity["label"] == "TACTIC":
            tactics.append(entity["word"])
        elif entity["label"] == "TECHNIQUE":
            techniques.append(entity["word"])
        elif entity["label"] == "PROCEDURE":
            procedures.append(entity["word"])

    return tactics, techniques, procedures

# Example usage
#text = "This is an example sentence. It contains some tactics, techniques, and procedures."
tactics, techniques, procedures = perform_ner(preprocessed_text)

print("Tactics:", tactics)
print("Techniques:", techniques)
print("Procedures:", procedures)
```

```python
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification

def extract_keywords(text):
    # Load the pre-trained model and tokenizer
    model_name = "dslim/bert-base-NER"
    #model_name = 'distilbert-base-uncased'  # Example model, change it to the one you want to use
    model = AutoModelForSequenceClassification.from_pretrained(model_name)
    tokenizer = AutoTokenizer.from_pretrained(model_name)

    # Tokenize the input text
    encoded_input = tokenizer(text, return_tensors='pt')

    # Pass the input through the model
    with torch.no_grad():
        output = model(**encoded_input)

    # Get the output embeddings
    embeddings = output.last_hidden_state.squeeze(0)  # Assuming batch size of 1

    # Calculate importance scores for each token
    importance_scores = embeddings.sum(dim=0)  # Sum the embeddings along the sequence length

    # Sort the tokens by importance scores
    sorted_indices = importance_scores.argsort(descending=True)

    # Convert token indices back to words
    keywords = [tokenizer.decode([idx.item()]) for idx in sorted_indices]

    return keywords
```

```python
import torch
from transformers import AutoTokenizer, AutoModelForTokenClassification

def extract_keywords_and_ner(text):
    # Load the pre-trained model and tokenizer
    model_name = 'bert-base-uncased'
    model = AutoModelForTokenClassification.from_pretrained(model_name)
    tokenizer = AutoTokenizer.from_pretrained(model_name)

    # Define the label mapping for NER
    label_map = {
        0: "O",
        1: "B-ENTITY",
        2: "I-ENTITY"
    }

    # Tokenize the input text
    encoded_input = tokenizer(text, return_tensors='pt')

    # Pass the input through the model
    with torch.no_grad():
        output = model(**encoded_input)

    # Get the predicted labels (NER)
    predicted_labels = torch.argmax(output.logits, dim=2)
    predicted_labels = predicted_labels.squeeze(0).tolist()

    # Get the tokens and convert labels to named entities
    tokens = tokenizer.convert_ids_to_tokens(encoded_input.input_ids.squeeze())
    entities = []
    entity = ""
    prev_label = None
    for token, label in zip(tokens, predicted_labels):
        if label == 0:  # O label (not part of any entity)
            if entity:
                entities.append(entity.strip())
                entity = ""
        elif label == 1:  # B-ENTITY label (beginning of an entity)
            if entity:
                entities.append(entity.strip())
            entity = token
        elif label == 2:  # I-ENTITY label (inside an entity)
            entity += " " + token

    if entity:
        entities.append(entity.strip())

    # Extract keywords
    keywords = [token for token, label in zip(tokens, predicted_labels) if label != 0]

    return keywords, entities
```
