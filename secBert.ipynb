```python
from nltk.tokenize import sent_tokenize, word_tokenize

def split_into_sentences(text):
    # Split text into sentences
    sentences = sent_tokenize(text)
    return sentences
```

```python
from transformers import AutoTokenizer, AutoModelForTokenClassification
import torch

tokenizer = AutoTokenizer.from_pretrained("jackaduma/SecBERT")
model = AutoModelForTokenClassification.from_pretrained("jackaduma/SecBERT")

def extract_ttps(sentence):
    inputs = tokenizer.encode(sentence, return_tensors="pt")
    outputs = model(inputs).logits
    predictions = torch.argmax(outputs, dim=2)

    ttps = []
    for token, label in zip(inputs[0], predictions[0]):
        if label == 1:
            ttps.append(tokenizer.decode([token]))
    return ttps
```

```python
from transformers import AutoTokenizer, AutoModelForTokenClassification
import torch

tokenizer = AutoTokenizer.from_pretrained("jackaduma/SecBERT")
model = AutoModelForTokenClassification.from_pretrained("jackaduma/SecBERT")

def extract_ttps(sentence):
    inputs = tokenizer.encode(sentence, return_tensors="pt")
    outputs = model(inputs).logits
    predictions = torch.argmax(outputs, dim=2)

    ttps = {"tactics": [], "techniques": [], "procedures": []}
    for token, label in zip(inputs[0], predictions[0]):
        if label == 1:
            ttps["tactics"].append(tokenizer.decode([token]))
        elif label == 2:
            ttps["techniques"].append(tokenizer.decode([token]))
        elif label == 3:
            ttps["procedures"].append(tokenizer.decode([token]))

    return ttps
```

```python
result=[]
for sentence in sentences[:5]:
  result.append(extract_ttps(sentence))
```

```python
for elem in result:
  print(elem)
```
